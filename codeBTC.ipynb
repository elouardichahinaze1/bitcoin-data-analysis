```python
import yfinance as yf 
import matplotlib.pyplot as plt 



data = yf.download('BTC-USD', start='2025-01-01')
print(data.head())


data['Close'].plot(figsize=(10, 5), color='purple')
plt.title('Bitcoin Price Trend - 2025')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True) 
plt.show()

data['SMA20'] = data['Close'].rolling(window=20).mean()
data['SMA50'] = data['Close'].rolling(window=50).mean()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(data['Close'], label='Price', color='black', alpha=0.7)
plt.plot(data['SMA20'], label='SMA 20', color='blue')
plt.plot(data['SMA50'], label='SMA 50', color='red')
plt.title('Stock Price with Moving Averages')
plt.legend()
plt.show()
# --- Suite du projet : Analyse de la Volatilité ---

# 1. Calcul des rendements quotidiens (Daily Returns)
# Cela calcule la variation en pourcentage d'un jour à l'autre
data['Daily Return'] = data['Close'].pct_change()

# 2. Visualisation des rendements
plt.figure(figsize=(12, 4))
data['Daily Return'].plot(color='green', alpha=0.8)
plt.title('Rendements Quotidiens du Bitcoin en 2025')
plt.axhline(0, color='black', lw=1) # Ligne à zéro pour voir les gains vs pertes
plt.ylabel('Variation %')
plt.show()

# 3. Histogramme de la distribution des rendements
# Utile pour voir si les variations sont extrêmes ou stables
plt.figure(figsize=(10, 5))
data['Daily Return'].hist(bins=50, color='orange', edgecolor='white')
plt.title('Distribution des Rendements (Volatilité)')
plt.xlabel('Variation quotidienne')
plt.ylabel('Fréquence')
plt.show()

# 4. Quelques statistiques clés avec Pandas
print(f"Rendement moyen : {data['Daily Return'].mean():.4f}")
print(f"Volatilité (Écart-type) : {data['Daily Return'].std():.4f}")
```
